```python
#Modules needed for compilation and execution
from IPython.display import clear_output
import random

#To print the TIC-TAC-TOE board
def display_board(board):
    clear_output()
    print(board[1] + " | " + board[2] + " | " + board[3])
    print(board[4] + " | " + board[5] + " | " + board[6])
    print(board[7] + " | " + board[8] + " | " + board[9])

#To accept player markers
def player_input():
    marker = ''
    while marker != 'X' and marker != 'O':
        marker = input("Player 1: Enter X or O ").upper()
    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')

#To define the position of marker
def place_marker(board, marker, position):
    board[position] = marker

#ALl possibilities for winning
def win_check(board, mark):
    return (
        (board[1] == mark and board[2] == mark and board[3] == mark) or
        (board[4] == mark and board[5] == mark and board[6] == mark) or
        (board[7] == mark and board[8] == mark and board[9] == mark) or
        (board[1] == mark and board[4] == mark and board[7] == mark) or
        (board[2] == mark and board[5] == mark and board[8] == mark) or
        (board[3] == mark and board[6] == mark and board[9] == mark) or
        (board[1] == mark and board[5] == mark and board[9] == mark) or
        (board[3] == mark and board[5] == mark and board[7] == mark)
    )

#Who goes first
def choose_first():
    flip = random.randint(0, 1)
    if flip == 0:
        first_player='Player1'
    elif flip==1:
        first_player='Player2'
    
    print("Start the game: ",first_player)
    return first_player
        
#To check for empty spaces on the board    
def space_check(board, position):
    return board[position] == ''

#To check if board is full
def full_check(board):
    for i in range(1, 10):
        if space_check(board, i):
            return False
    return True

#To choose player position
def player_choice(board):
    position = 0
    while position not in [1, 2, 3, 4, 5, 6, 7, 8, 9] or not space_check(board, position):
        try:
            position = int(input("Enter the position to place your marker: (1-9)"))
        except:
            print("Enter a valid number !")
    return position

#Restart game
def replay():
    choice = input("Play again? Yes or No ")
    return choice.upper() == 'YES'

#MAIN CODE
print("Welcome to Tic-Tac-Toe")
while True:
    the_board = [''] * 10
    player1_marker, player2_marker = player_input()
    turn = choose_first()
    game_on = True

    while game_on:
        if turn == 'Player1':
            display_board(the_board)
            position = player_choice(the_board)
            place_marker(the_board, player1_marker, position)
            if win_check(the_board, player1_marker):
                display_board(the_board)
                print("Player 1 has won!!!")
                game_on = False
            else:
                if full_check(the_board):
                    display_board(the_board)
                    print("It is a TIE!!")
                    break
                else:
                    turn = 'Player2'
        elif turn == 'Player2':
            display_board(the_board)
            position = player_choice(the_board)
            place_marker(the_board, player2_marker, position)
            if win_check(the_board, player2_marker):
                display_board(the_board)
                print("Player 2 has won!!!")
                game_on = False
            else:
                if full_check(the_board):
                    display_board(the_board)
                    print("It is a TIE!!")
                    break
                else:
                    turn = "Player1"

    if not replay():
        break
```
