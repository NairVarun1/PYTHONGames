```python
import random

def what_player_choice():
    while True:
        player_choice=input("Enter rock, paper or scissors ??").lower()
        if player_choice in ["rock","paper","scissors"]:
            return player_choice
        else:
            print("Please enter a valid choice")

def choose_winner(player_choice,comp_choice):
    if player_choice==comp_choice:
        return "tie"
    elif(
    (player_choice=="rock" and comp_choice=="scissors") or
    (player_choice=="paper" and comp_choice=="rock") or
    (player_choice=="scissors" and comp_choice=="paper")):
        return "player"
    else:
        return 'computer'

def print_score(player_count,comp_count):
    print(f" Player Score: {player_count} \n Computer Score: {comp_count}")
    
def game():
    choices=["rock","paper","scissors"]
    player_count=0
    comp_count=0
    rounds=3
    
    print("Welcome to Rock, Paper and Scissors !!")
    print("The WINNER will be decided on the basis of a Best of 3 Round score !! ")
    for _ in range(rounds):
        comp_choice=random.choice(choices)
        player_choice=what_player_choice()
    
        print(f"The Computer picked : {comp_choice}")
    
        winner=choose_winner(player_choice,comp_choice)
          
        if winner=='tie':
            print("It's a Tie")
        elif winner=='player':
            print("You WIN this round !")
            player_count+=1
        else:
            print("Computer WINS this round :|")
            comp_count+=1
              
        print_score(player_count,comp_count)
        print("-------------------------")
        
    if player_count>comp_count:
        print("WOOHOO You WIN the game !!!")
    elif comp_count>player_count:
        print("OOPS you Lose!! Better luck next time :D")
    else:
        print("TOO CLOSE!! IT's a TIE")

    
if __name__=="__main__":
    game()
        
        
        
```
